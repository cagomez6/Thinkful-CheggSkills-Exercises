```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy import stats
```

```python
PATH = 'https://downloads.ctfassets.net/c7lxnbtvvcxm/3LisiybW0TfzcWyxhmMMCq/0dc5bb212dc6f97a6f9aa41f81e39796/DS_-_ab_edited.csv'
data = pd.read_csv(PATH)
```

```python
data.info()
data.head()
```

1.Check for adequate sample sizes.
2.Check for changes over time in results.
3.Formally test for a significant difference in conversions between treatment and control groups.

```python
test = len(data[data.group == "treatment"])
control = len(data[data.group == "control"])
print('test sample size:', test)
print('control sample size:', control)
print('test proportion:', data.converted.mean())
# print()

print('test signup rate:', data[data.group == 'treatment'].converted.mean())
print('control signup rate:', data[data.group == 'control'].converted.mean())

stats.ttest_ind(data[data.group == 'treatment'].converted,
                data[data.group == 'control'].converted)
```

```python
data['date'] = pd.DatetimeIndex(data['timestamp']).date

fig = data[data.group=='treatment'][
    ['date', 'converted']].groupby('date').mean().plot()
data[data.group=='control'][
    ['date', 'converted']].groupby('date').mean().plot(ax=fig)
plt.legend(['treatment','control'])
plt.title('Comparing Conversion Rates by Treatment Date')
plt.show()
```

```python
print('T-Test Results by Date')

for day in data.date.unique():
    dated_data = data[data.date == day]
    print(stats.ttest_ind(dated_data[dated_data.group == 'treatment'].converted,
                          dated_data[dated_data.group == 'control'].converted))
```
